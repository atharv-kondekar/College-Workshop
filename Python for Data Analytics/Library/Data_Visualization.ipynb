```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([12,24,31,45,50,55,67])
y=np.array([10,23,29,34,39,49,60])
z=np.array([65,34,45,67,78,92,76])

plt.plot(x,y)
# x - on x axis
# y - on y axis
```

```python
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot(x,y,marker='o',mfc='Black')  # marker plot colour
```

```python
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot(x,y,marker='o',mfc='Black')  # marker plot colour
plt.plot(x,z,marker='o',mfc='orange')  # marker plot colour
```

```python
plt.title("Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot(x,y,marker='o',mfc='Black',label='2024'); # marker plot colour
plt.plot(x,z,marker='o',mfc='orange',label='2025')  # marker plot colour
plt.grid()
plt.legend();
```

# Box Plot 

```python
df4=pd.read_csv(r'C:\Users\athar\Downloads\Social_Network_Ads.csv')
df4
```

```python
plt.boxplot(df4['Age'])
df4.iloc[2,2]=2
```

```python
plt.boxplot(df4['Age'])
df4.iloc[78,2]=150 #outlyer
```

```python
q3=np.percentile(df4['Age'],75)
```

```python
q1=np.percentile(df4['Age'],55)
```

```python
iqr=q3-q1
iqr
```

```python
ub=q3+1.5*iqr #upper bound
ub
```

```python
lb=q1-1.5*iqr # lower bound
lb
```

```python
# any value above ub and below lb is considered as outlyer

df4[(df4['Age']>lb) & (df4['Age']<ub)]
```
